```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4]) # y = x+1
```

```python
plt.plot([1,2,3,4],[1,4,9,16]) # y = x^2
```

```python
font1 = {'family':'fantasy','color':'deeppink','weight':'normal','size':'xx-large'}
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('x',labelpad=30, loc='left', fontdict=font1)
plt.ylabel('y',loc='top')
```

```python
#handles  line1 (,)  .
line1 = plt.plot([1, 2, 3], label='label1')
line2 = plt.plot([1, 4, 9], label='label2')
line3 = plt.plot([1, 8, 27], label='label3')
# plt.legend(handles=[line1, line2], loc='best') #best best 
plt.legend(ncol=3,fontsize=10,frameon=True,shadow=True) #frameon False   
plt.title('legend test')
plt.grid(True,axis='both')
plt.show()
```

```python
import numpy as np
x = np.arange(3)
y = [100,400,900]
plt.bar(x,[100,400,900], color=['r','g','b'],width=0.7)
plt.xticks(x,['2018','2019','2020'])
plt.plot(x,y)
# plt.yticks(y,['a','b','c'])
plt.show()
```

```python
#  
import numpy as np
y = np.arange(3)
x = [100,400,900]
plt.barh(y,x,color=['r','g','b'],height=0.7)
plt.plot(x,y)
plt.yticks(y,['2018','2019','2020'])
plt.show()
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

area = (30*np.random.rand(n))**2
# plt.plot(x,y)
colors = np.random.rand(n)

plt.scatter(x,y,area,colors,cmap='Spectral',alpha=0.5)
```

```python
 # %%  %     
ratio = [24, 28, 12, 36]
labels = ['A','B','C','D']
wedgeprops = {'width':0.7,'edgecolor':'k'}
plt.pie(ratio, labels=labels,wedgeprops=wedgeprops,autopct='%.1f%%', explode=[0,0.05,0,0.2], counterclock=False,startangle=90)
```

```python
#1)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

line1 = plt.plot(months,sales_2019, label='2019')
line2 = plt.plot(months,sales_2020, label='2020')

plt.legend(fontsize=10)
plt.title('Monthly Sales Comparison (2019-2020)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
#2)
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20,35,15,27,45]
x = categories
y = data

plt.bar(x,y)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True,axis='both')
plt.xticks(rotation=45)
plt.ylim(0,50)

plt.show()
```

```python
# #3
labels = ['Apple','Banana','Mango','Blueberry']
sizes = [15,30,45,10]
explode = (0,0.1,0,0)
plt.pie(sizes, labels=labels,autopct='%.1f%%', explode=explode,startangle=90,shadow=True)
```

```python
#4)
x = np.linspace(0, 10, 100)

# sin cos   
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sin,label='sin(x)',color='b')
plt.plot(x,y_cos,label='cos(x)',color='r')

plt.title("Sine and Cosine Functions")
plt.xlabel("X")
plt.xlabel("Y")
plt.grid(True,axis='both')
plt.legend()

# plt.xlim(0,10)
# plt.ylim(-1,1)
```
